```python
from scipy.io import loadmat
import numpy as np
data = loadmat('sat-4-full.mat')
```

```python
trainx = data['train_x']
trainy = data['train_y']
```

## Data Preprocessing

```python
# The goal is to build a Data frame, which uses the image as unit. There are 784 pixels to describe each unit and four channels (red, green, blue and NIR components) to describe each pixel.

x=np.zeros(((4000,784,4))) 
w=0;
for l in range(0,4000):
    for i in range(0,28):
        for j in range(0,28):
            for k in range(0,4):
                x[l][w][k]=trainx[i][j][k][l]
            w=w+1
    w=0
```

```python
x
```

```python
# transfer the RGB to HSV

import colorsys

v=np.zeros(((4000,784,4)))
w=0
for l in range(0,4000):
    for w in range(0,784):
        r=x[l][w][0]
        g=x[l][w][1]
        b=x[l][w][2]
        v[l][w][0]=colorsys.rgb_to_hsv(float(r)/float(255), float(g)/float(255), float(b)/float(255))[0]
        v[l][w][1]=colorsys.rgb_to_hsv(float(r)/float(255), float(g)/float(255), float(b)/float(255))[1]
        v[l][w][2]=colorsys.rgb_to_hsv(float(r)/float(255), float(g)/float(255), float(b)/float(255))[2]
        v[l][w][3]=x[l][w][3]
```

```python
v
```

```python
# transfer the original data set to a specific result of each image (0-barren land; 1-trees; 2-grassland; 3-none).
n=[]
for i in range(0,4000):
	for j in range(0,4):
		if trainy[j][i] != 0:
			n.append(j)
```

```python
# calculate the mean value and standard deviation value of H, S and V separately in each image.

f=np.zeros((4000,9))
h=[]
s=[]
q=[]
m=[]
for l in range(0,4000):
    for w in range(0,784):
        h.append(v[l][w][0])
        s.append(v[l][w][1])
        q.append(v[l][w][2])
        m.append(v[l][w][3])
    f[l][0]=np.mean(h)
    f[l][1]=np.std(h)
    f[l][2]=np.mean(s)
    f[l][3]=np.std(s)
    f[l][4]=np.mean(q)
    f[l][5]=np.std(q)
    f[l][6]=np.mean(m)
    f[l][7]=np.std(m)
    f[l][8]=n[l]
    h=[]
    s=[]
    q=[]
    m=[]
```

```python
f
```

```python
import pandas as pd

# create the pandas DataFrame
df = pd.DataFrame(f, columns = ['hue_mean', 'hue_std', 'saturation_mean', 'saturation_std', 'value_mean', 'value_std', 'NIR_mean', 'NIR_std', 'land_type'])
# print dataframe
df
```

```python
import pandas as pd
from sklearn import preprocessing

x = df.iloc[:,6:8].values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)

df.iloc[:,6:8] = x_scaled

# 参考：https://medium.com/dunder-data/selecting-subsets-of-data-in-pandas-6fcd0170be9c
```

```python
df
```

## Parameter tuning - Cross validation in training data

```python
X=df.iloc[:,:8]
y=df.iloc[:,8]
```

```python
from sklearn import svm, datasets
from sklearn.model_selection import GridSearchCV
param_grid = [
  {'C': [0.1, 1, 10, 100, 1000, 10000], 'gamma': [0.01, 0.1, 1, 10, 100, 1000], 'kernel': ['rbf']},
 ]
svc = svm.SVC()
grid = GridSearchCV(estimator= svc,
                    param_grid = param_grid,
                    scoring = 'accuracy',
                    )

grid_results = grid.fit(X, y)

print('Best Score: ', grid_results.best_score_)
print('Best Params: ', grid_results.best_params_)

# cv_results_参考： https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html

# SVM_gridsearch 参考： https://towardsdatascience.com/hyperparameter-tuning-c5619e7e6624
# randomforest_gridsearch 参考： https://www.programcreek.com/python/example/104786/sklearn.grid_search.GridSearchCV
# scoring 参考： https://www.jianshu.com/p/d206318b98d4
# grid.fit().grid_scores 参考： https://www.jianshu.com/p/d206318b98d4
```

```python
grid_results.cv_results_
```

```python
# Heat Map
scores = grid_results.cv_results_['mean_test_score'].reshape(6,6)

param_C = [0.1, 1, 10, 100, 1000, 10000]
param_gamma = [0.01, 0.1, 1, 10, 100, 1000]

# heat_plot参考： https://qiita.com/bmj0114/items/8009f282c99b77780563
```

The goal of cross-validation is to define a data set to test the model in the training phase (i.e. validation data set) in order to limit problems like overfitting,underfitting and get an insight on how the model will generalize to an independent data set. It is important the validation and the training set to be drawn from the same distribution otherwise it would make things worse.

参考： https://towardsdatascience.com/cross-validation-70289113a072

```python
# optimal C
param_grid_C = [
  {'C': [1, 2, 4, 6, 8, 10, 20, 30, 40, 50, 100], 'gamma': [10], 'kernel': ['rbf']},
 ]
svc = svm.SVC()
grid_C = GridSearchCV(estimator= svc,
                    param_grid = param_grid_C,
                    scoring = 'accuracy',
                    )

grid_C_results = grid_C.fit(X, y)

print('Best Score: ', grid_C_results.best_score_)
print('Best Params: ', grid_C_results.best_params_)
```
